```python
import pandas as pd

# Remplacez 'file_path.csv' par le chemin d'accès réel à votre fichier CSV
file_path = "../data/academicScores.csv"

# Utilisez le délimiteur correct et ignorez la première ligne si elle contient des en-têtes
data = pd.read_csv(file_path, delimiter=';', skiprows=1, header=None)

# Supprimer les guillemets doubles
data = data.apply(lambda x: x.str.replace('"', '', regex=True))

# Remplacer ';' avec ','
data = data.apply(lambda x: x.str.replace(';', ','))

# Diviser la ligne en colonnes séparées
data = data[0].str.split(',', expand=True)

# Renommer les colonnes
data.columns = ["Index0","Index1", "Gender", "Race/Ethnicity", "Parental Level of Education",
                "Lunch", "Test Preparation Course", "Math Score",
                "Reading Score", "Writing Score"]

# Afficher le DataFrame résultant
data
```

```python
## Pour supprimer les deux premier columns que j'ai nommé index0 et index1

data.drop(columns=["Index0", "Index1"], inplace=True)
```

```python
data
```

```python
data = pd.DataFrame(data)

# Remplacer la colonne "Gender" par des variables binaires
data['Gender'] = data['Gender'].replace({'male': 0, 'female': 1})

data
```

```python
# Créer une série contenant les comptages de chaque catégorie
race_counts = data['Race/Ethnicity'].value_counts()

# Créer un diagramme circulaire
plt.figure(figsize=(8, 8))
plt.pie(race_counts, labels=race_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Répartition par Race/Ethnicité')
plt.axis('equal')  # Assure un aspect circulaire
plt.show()
```

```python
object_col =['Race/Ethnicity', 'Parental Level of Education',]
data = pd.get_dummies(data, columns = object_col)
data
```

```python
## Pour convertir les variables numérique qui était en format object sout format int
for columns in data.columns[5:]:
  data[columns]=data[columns].astype("int64")
```

```python
# Remplacer les catégories de "Lunch" par des valeurs binaires
lunch_mapping = {'standard': 0, 'free/reduced': 1}
data['Lunch'] = data['Lunch'].replace(lunch_mapping)
```

```python
data
```

```python
# Remplacer les catégories de "Test Preparation Course" par des valeurs binaires
course_mapping = {'none': 0, 'completed': 1}
data['Test Preparation Course'] = data['Test Preparation Course'].replace(course_mapping)
```

```python
# Regression linéaire
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
X = data.drop(columns="Writing Score")
y = data["Writing Score"]

print(X.shape, y.shape)
```

```python
X_train, X_eval, y_train, y_eval = train_test_split(X, y,test_size=0.2, shuffle=True,random_state=0)
```

```python
def evaluate_model(model):
    print("Model name: ", type(model).__name__)
    print("Model parameters: ", model.get_params())

    # Printing model accuracy
    model_train_mae = mean_absolute_error(y_train,model.predict(X_train))
    model_test_mae = mean_absolute_error(y_eval,model.predict(X_eval))

    print("Model Mean Absolute error on the train set : %.2f" % model_train_mae)
    print("Model Mean Absolute error on the test set : %.2f" % model_test_mae)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
evaluate_model(model)
```

```python
from sklearn.dummy import DummyRegressor
```

```python
dummy_model = DummyRegressor(strategy='mean')
dummy_model.fit(X_train, y_train)

evaluate_model(dummy_model)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(n_jobs=-1,max_samples=200)
rf_model.fit(X_train, y_train)
evaluate_model(rf_model)
```
