```python
%matplotlib inline
import numpy as np
from math import exp
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
fig=plt.figure()
ax=plt.axes()
plt.title("Mon premier")
plt.axis([-11,11,0,1500])
x=range(-10,11)
y=[exp(abs(i)) for i in x]
plt.plot(x,y,color="green" ,linestyle="solid",label="green")
plt.plot(x,[exp(abs(i+1)) for i in x],color="yellow",linestyle="solid",label="yellow")
plt.plot(x,[exp(abs(i-1)) for i in x],color="red",linestyle="solid",label="red")
plt.plot(x,[exp(abs(i+2)) for i in x],color="blue",linestyle="solid",label="blue")
plt.plot(x,[exp(abs(i+3)) for i in x])
plt.legend(loc="lower right")
```

```python
from math import sin
fig=plt.figure(figsize=(12,12))
x=range(-5,5)
for i in range(2) :
    fig.add_subplot(1,2,i+1)
    plt.style.use(plt.style.available[i])
    plt.scatter(x,[sin(j-i) for j in x])
    plt.text(s=plt.style.available[i],x=10,y=1,color="green")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

prets = pd.read_csv('https://raw.githubusercontent.com/OpenClassrooms-Student-Center/fr-4452741-decouvrez-les-librairies-python-pour-la-data-science/main/data/prets_final.csv')

prets.head()
```

```python
#nuage de point
plt.scatter(prets['revenu'] , prets['taux_endettement'] , color='green')
```

```python
#circulaire pour voir le nombre des clients en risque
clients_risque = prets.loc[prets['risque'] == 'Oui'].shape[0]

clients_non_risque = prets.shape[0] - clients_risque

plt.pie(x=[clients_risque , clients_non_risque] , labels=['En risque' , 'Pas de risque'])
```

```python
#En bar pour voir le nombre des clients en risque
plt.bar(height=[clients_risque , clients_non_risque] , x=['En risque' , 'Pas de risque'])
```

```python
#nuage de point pour tous les agence : le taux_interet en fonction du revenu

for ville in prets['ville'].unique() :
    df = prets.loc[prets['ville'] == ville , :]
    plt.scatter(df['revenu'] , df['taux_interet'] , label = ville)
    
plt.legend()
plt.show()
```
